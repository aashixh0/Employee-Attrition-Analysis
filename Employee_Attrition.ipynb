```python
def compare(df1, df2, index, column, values):
    dfp_leave = Attrition.pivot_table(index=index,
                                     columns=column,
                                     values=values, aggfunc=np.mean)
    dfp_stay = Stay.pivot_table(index=index,
                               columns=column,
                               values=values, aggfunc=np.mean)
    dfp = dfp_leave/dfp_stay
    return dfp
```
